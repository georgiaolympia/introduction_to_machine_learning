### Scikit-Learn

Scikit-Learn ist eine freie Software-Bibliothek zum maschinellen Lernen für die Programmiersprache Python. Es bietet verschiedene Klassifikations-, Regressions- und Clustering-Algorithmen, darunter Support-Vektor-Maschinen, Random Forest, Gradient Boosting, k-means und DBSCAN, und ist so konzipiert, dass es mit den numerischen und wissenschaftlichen Python-Bibliotheken NumPy und SciPy zusammenarbeitet.

Außerdem bietet Scikit-Learn eine Reihe von Standard Datensätzen an, z.B. den Iris Datensatz, den Digit Klassifikationsdatensatz, sowie den Boston House Prices Datensatz. Im folgenden werden wir mit diesen Datensäzten arbeiten.

http://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets

#### Datensätze aufrufen

```python
# Importiere das Modul datasets aus Scikit-Learn
from sklearn import datasets

# Lade den Iris Datensatz
# mehr Infos unter: 
# http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris
# http://scikit-learn.org/stable/datasets/index.html#iris-plants-database

iris = datasets.load_iris()

print("Iris Dataset Size:", iris.data.shape)

print("\n Iris:\n", iris.data)
```

```python
print("Iris Dataset Labels:\n", iris.target)
```

```python
# Visualisierung für zwei Dimensionen

import matplotlib.pyplot as plt

X = iris.data[:, :2]  # wir betrachten nur die ersten beiden Dimensionen des Datensatzes
y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```
