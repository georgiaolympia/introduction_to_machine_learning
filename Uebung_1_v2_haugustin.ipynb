### Scikit-Learn

Scikit-Learn ist eine freie Software-Bibliothek zum maschinellen Lernen für die Programmiersprache Python. Es bietet verschiedene Klassifikations-, Regressions- und Clustering-Algorithmen, darunter Support-Vektor-Maschinen, Random Forest, Gradient Boosting, k-means und DBSCAN, und ist so konzipiert, dass es mit den numerischen und wissenschaftlichen Python-Bibliotheken NumPy und SciPy zusammenarbeitet.

Außerdem bietet Scikit-Learn eine Reihe von Standard Datensätzen an, z.B. den Iris Datensatz, den Digit Klassifikationsdatensatz, sowie den Boston House Prices Datensatz. Im folgenden werden wir mit diesen Datensäzten arbeiten.

http://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets

#### Datensätze aufrufen

```python
# Importiere das Modul datasets aus Scikit-Learn
from sklearn import datasets

# Lade den Iris Datensatz
# mehr Infos unter: 
# http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris
# http://scikit-learn.org/stable/datasets/index.html#iris-plants-database

iris = datasets.load_iris()

print("Iris Dataset Size:", iris.data.shape)

print("\n Iris:\n", iris.data)
```

```python
print("Iris Dataset Labels:\n", iris.target)
```

```python
# Visualisierung für zwei Dimensionen

import matplotlib.pyplot as plt

X = iris.data[:, :2]  # wir betrachten nur die ersten beiden Dimensionen des Datensatzes
y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
# Rufe nun selbst den Diabetes Dataset auf
# mehr Infos unter: 
# http://scikit-learn.org/stable/datasets/index.html#diabetes-dataset
# http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes.html#sklearn.datasets.load_diabetes

diabetes = datasets.load_diabetes()

print(dir(datasets.load_diabetes()))

print(diabetes.feature_names)

# age, sex, body mass index, average blood pressure, and six blood serum measurements
print("Diabetes Dataset Size:", diabetes.data.shape)
print("\n Diabetes:\n", diabetes.data)

# Target is a quantitative measure of disease progression one year after baseline
print("Diabetes Dataset Target:\n", diabetes.target)
```

```python
# Visualisiere die Daten, sowie die Labels

import numpy as np
import pandas as pd

# Load as pandas DF (https://stackoverflow.com/questions/38105539/how-to-convert-a-scikit-learn-dataset-to-a-pandas-dataset)
diabetes_df = pd.DataFrame(data= np.c_[diabetes['data'], diabetes['target']], columns= diabetes['feature_names'] + ['target'])
diabetes_df.sample(5)
```

```python
# Create scatter matrix with pandas (https://stackoverflow.com/questions/7941207/is-there-a-function-to-make-scatterplot-matrices-in-matplotlib)
pd.plotting.scatter_matrix(diabetes_df, alpha = 0.2, figsize = (10, 10), diagonal = 'hist');
```

### Regression

Füher nun eine lineare Regression auf dem Diaetes Datensatz durch, die vorhersagt, wie weit die Krankheit bei einem Patienten vortgeschritten ist. Betrachte zunächste nur eine Variable (wie Alter oder BMI) und versuche anschließend eine multivariate Regression mit allen zur Verfügung stehenden Informationen zu implementieren.

Scikit-Learn stellt in seinem Modul für [lineare Modelle](http://scikit-learn.org/stable/modules/linear_model.html) eine Klasse für Lineare Regression bereit. Nutze diese. 

http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression

Beispielcode für eine eindimensionale Regression: http://scikit-learn.org/stable/auto_examples/linear_model/plot_ols.html#sphx-glr-auto-examples-linear-model-plot-ols-py

Teile das Datenset für die Evaluation in ein Trainings- und Testdatenset mit einem Verhältnis von 75:25 ein. Nutze hierfür das Modul [Model Selection](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.model_selection) mit der Train-Test-Split Klasse: http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

```python
# Aufteilung des Datensets in Trainings- und Testdatensatz

from sklearn import model_selection

X_train, X_test, y_train, y_test = model_selection.train_test_split(diabetes['data'], diabetes['target'], test_size=0.25)
print(X_train.shape, X_test.shape)
print(diabetes.data.shape)
```

```python
# Erzeuge eine Instanz der Linear Regression Klasse
# Nutze die Option normalized, um anzugeben, dass normalisierte Daten verwendet werden

from sklearn import linear_model

# Create linear regression object
regr = linear_model.LinearRegression(normalize=True)
```

```python
# Trainiere das Modell mit einem Input (z.B. Alter, BMI)
# mehr Infos: http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.fit
print('Alter: ',X_train[:,:1])

# Use only one feature
X_train_age = X_train[:,:1]
X_test_age = X_test[:,:1]

# Train the model using the training sets
regr.fit(X_train_age, y_train)
```

```python
# Evaluiere das Modell auf dem Testset
# Mache dafür zunächst eine Vorhersage für die Testdaten
# mehr Infos: http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression.predict
# und evaluiere anschließend die Genauigkeit des Modells

# Make predictions using the testing set
diabetes_y_pred = regr.predict(X_test_age)

# The coefficients
print('Coefficients: \n', regr.coef_)
print('Estimated intercept coefficient: \n', regr.intercept_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, diabetes_y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, diabetes_y_pred))
print('Score: ', regr.score(X_test_age, y_test))
```

```python
# Plot the training points

x_min, x_max = X_test_age[:, 0].min() - .05, X_test_age[:, 0].max() + .05
y_min, y_max = y_test.min() - 10, y_test.max() + 10

plt.scatter(X_train_age, y_train)
plt.xlabel('age')
plt.ylabel('diabetes')
plt.plot(X_test_age, diabetes_y_pred, color='black', linewidth=3)

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
# Erzeuge eine neue Instanz der Klasse Lineare Regression und trainiere diese auf allen Dimensionen des Datensatzes
# Evaluiere das Modell anschließen
from sklearn.metrics import mean_squared_error, r2_score

# Create linear regression object
regr_new = linear_model.LinearRegression(normalize=True)

# Train the model using the training sets
regr_new.fit(X_train, y_train)
diabetes_y_pred_new = regr_new.predict(X_test)

# The coefficients
print('Coefficients: \n', regr_new.coef_)
print('Estimated intercept coefficient: \n', regr_new.intercept_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, diabetes_y_pred_new))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, diabetes_y_pred_new))
print('Score: ', regr_new.score(X_test, y_test))
```

```python
plt.scatter(regr_new.predict(X_train), regr_new.predict(X_train) - y_train, c='b', s=40, alpha=0.5)
plt.scatter(regr_new.predict(X_test), regr_new.predict(X_test) - y_test, c='g', s=40)
plt.hlines(y = 0, xmin=0, xmax = 300)
plt.title('Residual Plot using training (blue) and test (green) data')
plt.ylabel('Residuals')
plt.show()
```
